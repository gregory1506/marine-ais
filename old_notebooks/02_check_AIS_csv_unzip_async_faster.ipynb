```python
import geopandas as gpd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile
```

```python
def make_dfs(zipfile:str) -> None:
    root = zipfile.split("\\AIS")[0] + "\\"
    tail = "AIS"+zipfile.split("\\AIS")[1].strip(" ")
    df = pd.read_csv(zipfile)
    #cargo vessels 
    cargo = df[(df.VesselType >= 70) & (df.VesselType < 80)].copy()
    cargo['BaseDateTime'] = pd.DatetimeIndex(cargo.BaseDateTime)
    #cargo vessels 
    tanker = df[(df.VesselType >= 80) & (df.VesselType < 90)].copy()
    tanker['BaseDateTime'] = pd.DatetimeIndex(tanker.BaseDateTime)
    return [cargo, tanker, root, tail]
def make_cargos(vessels, cargo):
    from multiprocessing import Pool
    import vesselfuncs
    from functools import partial
    f=partial(vesselfuncs.make_vessel, cargo)
    if __name__ ==  '__main__': 
        num_processors = 6
        p=Pool(processes = num_processors)
        output = p.map(f,[vessel for vessel in vessels])
    return pd.concat(output, ignore_index=True)
def make_tankers(vessels, tanker):
    from multiprocessing import Pool
    import vesselfuncs
    from functools import partial
    f=partial(vesselfuncs.make_vessel, tanker)
    if __name__ ==  '__main__': 
        num_processors = 6
        p=Pool(processes = num_processors)
        output = p.map(f,[vessel for vessel in vessels])
    return pd.concat(output, ignore_index=True)
```

```python
import os
for yr in ['2019','2020','2021']:
    list_of_files = [f"D:\\CUSPATIAL\\US_MARINE\\{yr}\\"+x for x in os.listdir(f"D:\\CUSPATIAL\\US_MARINE\\{yr}") if ".zip" in x]
    for zipfile in list_of_files:
        if zipfile >= "D:\\CUSPATIAL\\US_MARINE\\2020\\AIS_2020_07_02.zip":
            try:
                cargo, tanker, root, tail = make_dfs(zipfile)
                cvessels = [x for x in cargo.VesselName.unique()]
                tvessels = [x for x in tanker.VesselName.unique()]
                tdf = make_tankers(tvessels, tanker)
                cdf = make_tankers(cvessels, cargo)
                tdf.to_parquet(f"{root}/parquets/{tail[:-4]}_tanker_5min.parquet")
                cdf.to_parquet(f"{root}/parquets/{tail[:-4]}_cargo_5min.parquet")
                print(zipfile)
            except:
                continue
```
