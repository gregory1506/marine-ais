```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import LineString
import pydeck as pdk
```

```python
%%time
df = pd.read_parquet("trajectories/ALL_GOM_tanker_2019-2021_traj.parquet")
df = pd.read_json(df.to_json())
df
```

```python
#set of timestamps for slected MMSI
tss = sorted(list(set([x for y in df.loc[1:4].ts.tolist() for x in y if x > 0])))
```

```python
import time

layer = pdk.Layer(
    "TripsLayer",
    df.loc[1:4],
    get_path="path",
    get_timestamps="ts",
    get_color=[253, 128, 93],
    opacity=0.8,
    width_min_pixels=5,
    rounded=True,
    trail_length=100000,
    current_time=tss[0]
)
view_state = pdk.ViewState(latitude = 29.41129858, longitude = -94.8185302, zoom=4, bearing=20, pitch=20)
# Render
r = pdk.Deck(
    map_provider="mapbox",
    map_style=pdk.map_styles.SATELLITE,
    api_keys={"mapbox": "pk.eyJ1IjoiZ3JvbDIwMjAiLCJhIjoiY2tvajI0MW5pMDMxMDJ3bzdpN3dzbHBidyJ9.jxYSuCXT1u1eFKMMWwSFVg"},
    layers=[layer],
    initial_view_state=view_state
)
r.show()
```

```python
# Animate
for i, ct in enumerate(tss):
    layer.current_time = ct
    r.update()
    time.sleep(0.05)
    if i % 100 == 0:
        print(i, ct)
```
