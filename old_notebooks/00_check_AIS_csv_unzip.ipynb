```python
import geopandas as gpd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import zipfile
```

```python
import os
list_of_files = [x for x in os.listdir("D:\\CUSPATIAL\\US_MARINE\\2021") if ".zip" in x]
list_of_files[-5:]
```

```python
# zfile = list_of_files[1]
for zfile in list_of_files:
    df = pd.read_csv(f"../US_MARINE/2021/{zfile}")
    #cargo vessels 
    cargo = df[(df.VesselType >= 70) & (df.VesselType < 80)].copy()
    cargo['BaseDateTime'] = pd.DatetimeIndex(cargo.BaseDateTime)
    #cargo vessels 
    tanker = df[(df.VesselType >= 80) & (df.VesselType < 90)].copy()
    tanker['BaseDateTime'] = pd.DatetimeIndex(tanker.BaseDateTime)
    list_of_cargo = []
    for i, vessel in enumerate(cargo.VesselName.unique()):
        tmp = cargo[cargo.VesselName == vessel].set_index('BaseDateTime').resample('5T').max().copy().dropna()
        tmp['VesselName'] = vessel
        tmp = tmp.reset_index(drop=False)
        list_of_cargo.append(tmp)
        if i % 100 == 0:
            print(f"Completed Vessel {i} {zfile} {vessel} cargo")
    cargo5m = pd.concat(list_of_cargo, ignore_index=True)
    cargo5m.to_parquet(f"../US_MARINE/2021/parquets/{zfile[:-4]}_cargo_5min.parquet")

    list_of_tanker = []
    for i, vessel2 in enumerate(tanker.VesselName.unique()):
        tmp2 = tanker[tanker.VesselName == vessel2].set_index('BaseDateTime').resample('5T').max().copy().dropna()
        tmp2['VesselName'] = vessel2
        tmp2 = tmp2.reset_index(drop=False)
        list_of_tanker.append(tmp2)
        if i % 100 == 0:
            print(f"Completed Vessel {i} {zfile} {vessel2} tanker")
    tanker5m = pd.concat(list_of_tanker, ignore_index=True)
    tanker5m.to_parquet(f"../US_MARINE/2021/parquets/{zfile[:-4]}_tanker_5min.parquet")
```
